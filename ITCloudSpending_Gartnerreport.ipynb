```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
table_MN = pd.read_html("https://www.gartner.com/en/newsroom/press-releases/2022-06-02-gartner-says-worldwide-iaas-public-cloud-services-market-grew-41-percent-in-2021",header=0)
```

```python
type(table_MN)
```

```python
df = table_MN[0]
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# Pie chart
labels = df['Company'].head(6)
sizes = df['2020-2021 Growth (%)'].head(6)
#explode = (0,0,0,0)


fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=None, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title('2020-2021 Growth (%)',loc ='center')


plt.show()
```

```python
import matplotlib.pyplot as plt

##fig1, ax1 = plt.subplots()
fig, ax1 = plt.subplots(figsize =(15, 5))

ax1.barh(df['Company'].head(6), df['2021 Revenue'].head(6))

ax1.invert_yaxis()

for i in ax1.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10,
             color ='black',fontweight ='bold')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize =(15, 5))

df = df[df.Company != 'Total']


n = df[df.columns[0]].count()
r = np.arange(n)
width = 0.2

plt.bar(r, df['2020 Revenue'], color = 'b',width = width, edgecolor = 'black',label=df['Company'])
plt.bar(r + width, df['2021 Revenue'], color = 'g',width = width, edgecolor = 'black',label=df['Company'])

plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Comparison of Revenue across companies in 2020-2021")
plt.xticks(r + width,df.Company)

plt.show()
```

```python
fig = plt.figure(figsize=(10,10),dpi=200)

ax1 = fig.add_subplot(121)

ax1.pie(df['2020 Market  Share (%)'], explode=None, labels=df['Company'], autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.set_title('2020 Market  Share (%)',loc ='center')


ax2 = fig.add_subplot(122)
ax2.pie(df['2020-2021 Growth (%)'], explode=None, labels=df['Company'], autopct='%1.1f%%',
        shadow=False, startangle=90)
ax2.set_title('2021 Market  Share (%)',loc ='center')


plt.show()
```
